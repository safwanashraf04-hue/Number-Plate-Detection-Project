### Import necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Create a function to display images in larger size

```python
def display(img):
    plt.figure(figsize=(10,8))
    plt.imshow(img)
```

```python
img=cv2.imread('car.jpg')
```

```python
display(img)
```

### Convert image from BGR to RGB format

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display(img)
```

### Loading the haarcascade_russian_plate_number.xml file

```python
plate_cascade=cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    plate_img=img.copy()
    plate_rects=plate_cascade.detectMultiScale(plate_img,minNeighbors=4)
    for x,y,w,h in plate_rects:
        cv2.rectangle(plate_img,(x,y),(x+w,y+h),(0,0,255),3)
    return plate_img
```

```python
result=detect_plate(img)
```

```python
display(result)
```

### Blurring license plate

```python
def detect_and_blur_plate(img):
    plate_img=img.copy()
    roi=img.copy()
    plate_rects=plate_cascade.detectMultiScale(plate_img,minNeighbors=4)
    for (x,y,w,h) in plate_rects:
        roi=roi[y:y+h,x:x+w]
        blurred_roi=cv2.medianBlur(roi,7)
        plate_img[y:y+h,x:x+w]=blurred_roi
    return plate_img
```

```python
result=detect_and_blur_plate(img)
```

```python
display(result)
```
